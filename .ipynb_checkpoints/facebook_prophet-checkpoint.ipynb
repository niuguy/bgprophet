```python
import os
import gzip
import json
import pandas as pd
import _pickle as pickle
import numpy as np
import matplotlib.pyplot as plt
import datetime
from fbprophet import Prophet
%matplotlib inline
```

```python
bg_path = '/home/jupyter/work/openaps/data/99908129/direct-sharing-31/entries_2017-01-01_to_2017-10-24.json'
df_bg = pd.read_json(bg_path)
```

```python
cols=['date', 'sgv','type']
df_bg=df_bg[cols]
df_bg = df_bg.set_index('date').sort_index()
```

```python
start='2017-06-01'
end='2017-06-10'
test_range = pd.date_range(start, end)
```

```python
m = Prophet()
df_bg['ds']=df_bg['date']
df_bg['y'] = df_bg['sgv']
m.fit(df_bg)
```

```python
future = m.make_future_dataframe(periods=10)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
df_bg[start:end].plot(y='sgv', figsize=(20, 6))
```

```python
for d in test_range:
    oneday_variation =  df_bg[str(d):str(d+pd.Timedelta(days=1))]
    if oneday_variation.empty:
        continue
    oneday_variation.plot(y='sgv')
```
